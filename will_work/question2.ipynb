```python
# Set up imports
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import seaborn as sns
from pathlib import Path
```

```python
# Import CSV
filepath = "Resources/student_health_data.csv"
dirty_data = pd.read_csv(filepath)

# Check it
dirty_data.info()
```

```python
dirty_data.head()
```

```python
clean_data = dirty_data
```

```python
# lets make integers
clean_data["Blood_Pressure_Systolic"] = clean_data["Blood_Pressure_Systolic"].astype("int")
clean_data["Heart_Rate"] = clean_data["Heart_Rate"].astype("int")
clean_data["Blood_Pressure_Diastolic"] = clean_data["Blood_Pressure_Diastolic"].astype("int")
clean_data["Stress_Level_Biosensor"] = clean_data["Stress_Level_Biosensor"].astype("int")
clean_data["Stress_Level_Self_Report"] = clean_data["Stress_Level_Self_Report"].astype("int")
clean_data["Study_Hours"] = clean_data["Study_Hours"].astype("int")
clean_data["Project_Hours"] = clean_data["Project_Hours"].astype("int")
clean_data.info()
```

```python
clean_data.head()
```

```python
df=clean_data
```

```python
# Question #2
# How does age and/or gender contribute to increased health risk.

# Age vs Health Risk Level

# Plot a boxplot to see how age affects the health risk level distribution.
# Boxplot to visualize age distribution By Health Risk Level 

# Define custom colors for each health risk level
risk_colors = {'Low': '#06C670', 'Moderate': '#2568E4', 'High': '#C55530'}

# Plot the boxplot with custom colors and set 'Health_Risk_Level' as the hue
plt.figure(figsize=(10, 6))
sns.boxplot(x='Health_Risk_Level', y='Age', data=df, hue='Health_Risk_Level', palette=risk_colors)
plt.title('Age vs Health Risk Level')  
plt.show()
```

```python
# Gender vs Health Risk Level 
# A Barplot can show the Distrubution of health risks between different genders 
gender_colors = {'M': 'lightblue', 'F': 'pink'}
sns.countplot(x='Health_Risk_Level', hue='Gender', data=df, palette=gender_colors)
plt.title('Gender Distribution by Health Risk Level')
plt.show()
```

```python
# Age and Gender combined effect on health risk level
gender_colors = {'M': 'lightblue', 'F': 'pink'}
plt.figure(figsize=(10, 6))
sns.violinplot(x='Health_Risk_Level', y='Age', hue='Gender', data=df, palette=gender_colors, split=True)
plt.title('Age and Gender Combined Effect on Health Risk Level')
plt.show() 
```
